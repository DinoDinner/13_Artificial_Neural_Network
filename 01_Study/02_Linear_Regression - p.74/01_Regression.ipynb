단순 선형 회귀의 개념 간단 실습

```python
import numpy as np
import matplotlib.pyplot as plt
```

예시 데이터 생성

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])
print(x)
print(y)
```

기울기와 절편 구하기

```python
beta_1 = np.sum((x-np.mean(x)) * (y-np.mean(y)))/np.sum((x-np.mean(x))**2)
beta_0 = np.mean(y) - beta_1 * np.mean(x)
```

결과 출력

```python
print('기울기 : ',beta_1)
print('절편 : ',beta_0)
```

```python
"""
y = 1.0x + 1.0
"""
# 산점도 그리기
plt.scatter(x,y)
plt.plot(x, beta_0 + beta_1 *x, color='red')
plt.show()
```
