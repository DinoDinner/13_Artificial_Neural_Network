평균 제곱 오차 간단한 실습

```python
import numpy as np

def mean_squred_error (y_true, y_pred) :
    mean_temp = np.mean((y_true - y_pred)**2)
    return mean_temp
"""
예시 y_true = [1,2,3,4] y_pred = [2,3,5,6]
"""
y_true = np.array([1,2,3,4])
y_pred = np.array([1,2,3,4])

mean_value = mean_squred_error(y_true,y_pred)
print (mean_value)
```

Binary Cross Entropy Loss 간단한 실습

```python
import numpy as np

def binary_cross_entropy_loss(y_true_BC, y_pred_BC):
    epsilon = 1e-5
    n_BC = y_true_BC.shape[0]
    print('n value >>', n_BC)
    
    loss = -np.mean(y_true_BC * np.log(y_pred_BC + epsilon) + (1 - y_true_BC) * np.log(1 - y_pred_BC + epsilon))
    return loss

y_true_BC = np.array([1, 0, 1, 1])
y_pred_BC = np.array([0.99, 0.1, 0.8, 0.7])

loss = binary_cross_entropy_loss(y_true_BC, y_pred_BC)
print(loss)
```

Cross Entropy Loss 간단한 실습

```python
import numpy as np

def cross_entropy_loss(y_true_C, y_pred_C, c):
    
    """
    y_true_C => 실제값 (크기 : (n,c))
    y_pred_C => 예측값 (크기 : (n,c))
    c        => 클래스 개수
    """
    
    
    n_C = y_true_C.shape[0]
    print('n value >>', n_C)
    ce_loss = -1 / n_C * np.sum(y_true_C * np.log(y_pred_C))
    
    return (ce_loss)

y_true_C = np.array([[1, 0, 1], [1, 1, 1], [0, 1, 1]])    # 실제값
y_pred_C = np.array([[0.99, 0.1, 0.8], [0.1, 0.8, 0.7], [0.99, 0.8, 0.7]])    # 예측값
c = 3

loss = cross_entropy_loss(y_true_C, y_pred_C, c)
print(loss)
```

인공지능 신경망 - 렐루 함수 실습

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x) :
    return np.maximum(0,x)

x = np.linspace(-5,5,1000)
y = relu(x)

plt.plot(x,y)
plt.show()
```

인공지능 신경망 - Leaky ReLU 함수 실습

```python
import matplotlib.pyplot as plt 
import numpy as np

def leaky_relu(x, alpha=0.01) :
    return np.where(x >= 0, x, x * alpha)

x_temp = np.linspace (-5, 5, 100)
y_tmep = leaky_relu(x_temp)
plt.plot(x_temp,y_tmep)
plt.show()
```
